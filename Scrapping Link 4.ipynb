```python
# from time import sleep
# from random import randint
from bs4 import BeautifulSoup
# import datetime
import requests
import pandas as pd
import re
```

```python
def get_page_source(url):
    agent = {"User-Agent":"Mozilla/5.0"} # User Agent defined to avoid robot detection
    source=requests.get(url, headers=agent)
    return source.text, source.status_code

def get_soup(content):
    soup=BeautifulSoup(content, "lxml")
    return soup
```

```python
def get_data_from_table(url = "https://fiscal.treasury.gov/surety-bonds/list-admitted-reinsurers.html"):
    text, status = get_page_source(url)
    soup = get_soup(text)
    
    result = list()
    table = soup.find('table')

    # HEADERS -- for Column names
    all_th = table.findAll('th')
    columns = []
    for th in all_th:
        columns.append(th.text.strip())

    rows = table.findAll('tr')
    for row in rows[1:]: # zero index has headings
        all_td = row.findAll('td')
        values = {}
        for i in range(len(columns)):
            values.update({columns[i] : all_td[i].text.strip()})

        result.append(values)
    return pd.DataFrame(result)
```

```python
df = get_data_from_table()
```

```python
df.to_csv("list_admitted_reinsurers.csv")
df
```
